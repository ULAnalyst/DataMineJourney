```python
#Group by 'Dest' and calculate average arrival delay - Groups the myAirplane_Avg DataFrame by destination (Dest), 
#calculates the average arrival delay (ArrDelay) for each destination, and resets the index to return a clean DataFrame
```

```python
pd.set_option('display.max_rows',None) # - pd.set_option(): A Pandas function used to configure global display options.
#display.max_columns: Specifies the option to control the maximum number of columns displayed.
#None: Sets the option to display all columns, regardless of how many there are
myAirplane_Avg.groupby('Dest')['ArrDelay'].mean().reset_index(name='Avg_Delay')
```
