```python
%%sql
SELECT s.playerID,
p.nameFirst,
p.nameLast,
p.nameGiven,
p.weight,
s.salary
FROM salaries as s
FULL JOIN people as p
ON s.playerID = p.playerID 
WHERE p.playerID like 'ruthba%'
LIMIT 5;

```

```python
%%sql
SELECT 
p.playerID,
s.salary

FROM salaries as s
JOIN people as p
ON s.playerID = p.playerID
 LIMIT 5;
```

SELECT statement chooses specific columns to display from the data tables.

Joins two tables: salaries and people so we can match each player’s ID with their salary using the playerID.

In the final part, it selects the playerID and their salary from the joined data.

Limits the list to only show the first 5 results.

```python
%%sql
WITH pSalary_data AS (
SELECT 
p.weight,
AVG(s.salary) as Salary

FROM salaries as s
JOIN people as p
ON s.playerID = p.playerID
GROUP BY p.weight
)

SELECT 
weight, 
Salary as avgSalary 
FROM pSalary_data 
ORDER BY Salary DESC 
LIMIT 10;
```

WITH statement creates temporary data called pSalary_data. It is a short-term table to be used later in the query.

Joins two tables: salaries and people so we can match each person’s weight with their salary using the playerID.

AVG function calculates the average salary for all players with the same weight.

Groups the results by weight so each weight value has one average salary.

In the final part, it selects each weight and their average salary from our temporary data.

Sorts the list so the players with the highest average salaries are at the top.

Limits the list to only show the top 10 results.

```python
%%sql
WITH pBatting_data AS (
    SELECT p.playerID, 
    sum(b.r) as totalPointsScored
    FROM people as p
    JOIN batting as b
    ON p.playerID = b.playerID
    GROUP BY p.playerID
    
)
select playerID, totalPointsScored FROM pBatting_data ORDER BY totalPointsScored DESC LIMIT 10;
```

```python
%%sql
SELECT cp.playerID, 
    sum(b.R) as totalRunScored
    FROM CollegePlaying as cp 
    JOIN Batting as b
    ON cp.playerID = b.playerID  
    WHERE cp.schoolID = 'purdue'
    GROUP BY b.R
    ORDER BY b.R DESC
    LIMIT 10;
```

```python
%%sql
WITH cpBatting_data AS (
    SELECT cp.playerID, 
    sum(b.R) as totalRunScored
    FROM CollegePlaying as cp 
    JOIN Batting as b
    ON cp.playerID = b.playerID  
    WHERE cp.schoolID = 'purdue'
    GROUP BY cp.playerID
)
Select 
playerID,
totalRunScored
From cpBatting_data
ORDER BY totalRunScored DESC
LIMIT 10;
```

WITH statement creates temporary data called cpBatting_data. It is a short-term table to be used later in the query.

Joins two tables: CollegePlaying and Batting so we can match each player with their batting records using the playerID.

Uses a WHERE condition to only include players from the school with the ID 'purdue'.

sum adds up the total number of runs (b.R) each player has scored, giving us totalRunScored.

Groups the results by playerID, so each player’s total runs are in one row.

```python
%%sql
WITH cpBatting_data AS (
    SELECT cp.playerID, 
    sum(b.R) as totalRunScored
    FROM CollegePlaying as cp 
    JOIN Batting as b
    ON cp.playerID = b.playerID  
    WHERE cp.schoolID = 'purdue'
    GROUP BY cp.playerID
)
Select 
playerID,
totalRunScored
From cpBatting_data
ORDER BY totalRunScored DESC
LIMIT 1;
```

```python
%%sql
WITH cpBatting_data AS (
    SELECT cp.schoolID, 
    AVG(b.R) as avgRunScored
    FROM CollegePlaying as cp 
    JOIN Batting as b
    ON cp.playerID = b.playerID  
    GROUP BY cp.schoolID
)
Select 
schoolID,
avgRunScored
From cpBatting_data
ORDER BY avgRunScored DESC
LIMIT 10;
```
