```python
# Create age categories using numerical bins
# pd.cut() categorizes ages into specified ranges
# value_counts() tallies records in each category
# sort_index() arranges categories in ascending order
```

```python
Passing_Category = pd.cut(myPassing_Rec['Age'], bins=[0,18,25,35,55,150,float("inf")]).value_counts().sort_index()
```

```python
# Create age categories with descriptive text labels
# Same binning as above but with human-readable labels
# Labels correspond to each bin interval defined above
```

```python
Passing_CategoryLabels = pd.cut(myPassing_Rec['Age'], bins=[0,18,25,35,55,150,float("inf")],labels=['youth','young adult','adult','middle age','senior','unknown']).value_counts().sort_index()
```

```python
# Convert numbered categories to DataFrame format
# reset_index() transforms the index (bin ranges) into a column
# name parameter renames the counts column for clarity
# Note: [0:18] is unnecessary since we only have 6 categories
```

```python
Passing_Category[0:18].reset_index(name='Age Group at Time of Passing')
```

```python
def makeatable (myseries: pd.Series) -> float:
    """
    The makeatable function takes a Pandas Series as input,
    with any type of data in the Series (for instance, it might be a column of a data frame)
    and returns the percentages of the data of each type.

    Args:
    myseries (pd.Series): This is any type of Series that we want to study,

    Returns:
    makeatable (pd.Series): This is a Pandas Series with the percentages of the data of each type.
    """
    mypercentages = myseries.value_counts(normalize = True)
    return mypercentages
```

```python
def teenagecount (myseries: pd.Series) -> float:
    """
    The teenagecount function takes a Pandas Series as input,
    with any type of data in the Series (for instance, it might be a column of a data frame)
    and returns the number of data points that are between 13 and 19 (inclusive).

    Args:
    myseries (pd.Series): This is any type of Series that we want to study,

    Returns:
    teenagecount (int): This is an integer that shows the number of values between 13 and 19 in the data.
    """
    teenagecount = len(myseries[(myseries >= 13) & (myseries <= 19)])
    return teenagecount
```
