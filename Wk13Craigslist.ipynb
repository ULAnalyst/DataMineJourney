```python
# Load the full dataset, but only the columns needed for geospatial analysis: state, longitude, and latitude
myClassifiedAll = pd.read_csv("/anvil/projects/tdm/data/craigslist/vehicles.csv", usecols =['state', 'long','lat'])
myClassifiedAll  # Display the DataFrame with only the selected columns to verify correct loading
```

```python
# Filter for vehicles listed in Indiana ("in") that have valid (non-null) longitude and latitude values
indyDF = (myClassifiedAll['state'] == "in") & (myClassifiedAll['long'].notnull()) & (myClassifiedAll['lat'].notnull())
indyDF  # This is a boolean Series showing which rows meet all filter criteria
```

```python
indyDF_filtered = myClassifiedAll[indyDF] # Create a DataFrame with only Indiana vehicles with valid coordinates
indyDF_filtered.head()  # Display the first few rows of the filtered DataFrame for verification
```

```python
# Convert the DataFrame to a GeoDataFrame, creating geometry points from longitude and latitude columns
gdf = gpd.GeoDataFrame(indyDF_filtered, geometry=gpd.points_from_xy(indyDF_filtered.long, indyDF_filtered.lat), crs="4326")  # get the longitudes and latitudes from indyDF_filtered
gdf = gdf.assign(mycolors='orange')   # make the Craigslist data orange
gdf.explore(color = gdf['mycolors'])
```
