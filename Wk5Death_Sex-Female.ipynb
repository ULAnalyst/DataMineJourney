```python
myPassing_RecNull=myPassing_Rec[myPassing_Rec['Sex'].notnull()]
```

```python
myPassing_RecNull.shape
```

```python
myFPassing_RecNull= myPassing_RecNull[(myPassing_RecNull['Sex'] ==  "F")]
```

```python
myFPassing_RecNull.shape
```

```python
myFPassing_RecNull.head()
```

```python
# Filter FEMALE records with non-null Sex values 
myFPassing_Rec = myPassing_Rec[(myPassing_Rec['Sex'] == "F") & (myPassing_Rec['Sex'].notnull())]
```

```python
# Create frequency distribution of ages at death for females
myFPassing_age_distribution = myFPassing_Rec['Age'].value_counts().sort_index()
```

```python
myFPassing_age_distribution = myFPassing_age_distribution.reset_index(name='Frequency_of_Death_at_Age')
```

```python
print(myFPassing_age_distribution)
```

```python
# Create bar plot using proper x and y values
plt.figure(figsize=(12, 6))
plt.bar(myFPassing_age_distribution['Age'].index, myFPassing_age_distribution['Frequency_of_Death_at_Age'])
plt.title('Frequency of Deaths by Age in Female Population')
plt.xlabel('Age')
plt.ylabel('Number of Deaths')
plt.show()
```

```python
# Filter FEMALE records where Age is not 999
myFPassing_999 = myFPassing_age_distribution(myFPassing_age_distribution['Age'].value_counts() != 999)
```
