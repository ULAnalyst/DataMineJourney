```python
def myavgstars (mybrand: str) -> float:
    """
    The myavgstars function takes a brand name as input,
    and returns the average number of stars for that brand.

    Args:
    mybrand (str): This is the name of the brand ("bj", "breyers", "hd", or "talenti") that we want to study.

    Returns:
    myavgstars (float): This is the average number of stars for the brand.
    """
    myRIData = pd.read_csv("/anvil/projects/tdm/data/icecream/combined/reviews.csv")
    myavgstars = myRIData[myRIData["brand"] == mybrand]["stars"].mean()
    return myavgstars
```
