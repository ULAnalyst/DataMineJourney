```python
#Group by 'date' and calculate mean "Avg_Score - Groups the beer reviews DataFrame by date, calculates the mean (Avg_Score) for the 3 dates with a score of 5,
# sort the values, call the last 3 rows of data, and resets the index and renames the calculated column as Avg_Score
```

```python
pd.set_option('display.max_rows',None)
myAvgBeer_Review=myBeer_Review.groupby('date')['score'].mean().sort_values().reset_index(name='Avg_Score')
#Groups the data by the date column (representing date)
#Averages the beer review scores to calculate (mean) the Avg_Score column for each date group to calculate Avg_Score by date
#Sorts the calculated avg score values in ascending order
#Selects the top 3 date with the highest Avg_Review that equals 5 by taking the last 3 rows of the sorted data
#Resets the index and renames the calculated column as Avg_Spend

```
