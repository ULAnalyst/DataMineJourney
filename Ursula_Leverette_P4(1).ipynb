```python
%%bash
# cell magic tells the kernal to expect a different language than the default Python.(TA Michael Hansen comments cannot be on the cell magic line)
# This command counts and prints the total number of orders for each gender group in the test_full.csv file.
# It uses awk with comma as the column separator (-F',').
# NR > 1 skips the header row (first row).
# count[$2]++ increments the count for the gender found in the second column.
# The END block prints each gender and its total order count.
cat /anvil/projects/tdm/data/restaurant/test_full.csv | awk -F',' 'NR > 1 {count[$2]++} END {for (gender in count) print gender, count[gender]}'
```
