```python
yearCount = mySports['Year'].nunique()
```

```python
# Print the number of times unique years appear in the data set
```

```python
# Count the occurences for each year
```

```python
yearCount = mySports['Year'].value_counts()
```

```python
# print the year and the number of occurences it appears in the dataset 
```

```python
print(yearCount)
```

```python
# import matplotlib library - is used for creating visualizations like bar charts.
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot a bar chart for illustrating year count
```

```python
plt.figure(figsize=(18, 8))
plt.bar(yearCount.index, yearCount, color='darkorange')  
plt.xlabel("Year")
plt.ylabel("Number of Records")
plt.title("The Frequency a Year appears in the Dataset")
plt.show()
```
