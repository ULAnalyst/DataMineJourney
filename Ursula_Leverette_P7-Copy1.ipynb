## Question 3

 - 3.2. Use HAVING to filter teams with an average salary greater than 7.5 million.

```python
%%sql 
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
GROUP BY t.name, s.yearID
ORDER BY avgSalary DESC
LIMIT 100;
```

```python
%%sql 
WITH tSalary_data AS (SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 7500000
ORDER BY avgSalary DESC
)

SELECT    
teamName,     
avgSalary  
FROM tSalary_data
```

Joins two tables: Salaries and Teams so we can match each team’s records with their salary information using teamID and yearID.

AVG calculates the average salary for each team, giving us avgSalary.

Groups the results by teamName so each team’s average salary is shown in one row.

HAVING filters the results to only include teams with an average salary greater than 7,500,000.

In the final part, it selects each teamName and their avgSalary from our temporary data.

Sorts the list so the teams with the highest average salaries appear first.

## Question 4

 - 4.1. A single SQL query returning teamName, yearID, avgSalary, and numpeoples.
 - 4.2. Use HAVING to filter teams with an average salary greater than 6.8 million and more than 20 peoples.
 - 4.3. Use WHERE to filter results for the years 2010 to 2015.

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numpeoples
FROM Salaries s
JOIN Teams t
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
GROUP BY t.name, s.yearID
ORDER BY avgSalary DESC
LIMIT 10;
```

SELECT statement chooses specific columns: the team name, year, average salary, and the number of players.

Joins two tables: Salaries and Teams, connecting them on matching teamID and yearID so each salary record aligns with the correct team and year.

Groups the results by team name and yearID so that each row shows the average salary and player count for that team in that year.

Orders the results by average salary in descending order, so teams paying the highest average salaries appear first.

Limits the output to the top 10 rows based on average salary.

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numpeoples
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 6800000 AND COUNT(s.playerID) > 20
ORDER BY avgSalary DESC
LIMIT 10;
```

SELECT statement chooses specific columns: the team name, year, average salary, and the number of players.

Joins two tables: Salaries and Teams on matching teamID and yearID, linking salary records to their correct team and year.

Groups the results by team name and yearID so that each row shows aggregated data for a team in a specific year.

HAVING clause filters to only include teams with an average salary greater than 6,800,000 and more than 20 players.

Orders results by average salary in descending order so that the highest-paid teams appear first.

Limits the output to the top 10 rows based on those conditions.

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numpeoples
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 6800000 AND COUNT(s.playerID) > 20
ORDER BY avgSalary DESC
LIMIT 10;
```

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numpeoples
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
WHERE s.yearID between 2010 and 2015
GROUP BY t.name, s.yearID
HAVING AVG(s.salary) > 6800000 AND COUNT(s.playerID) > 20
ORDER BY avgSalary DESC
LIMIT 10;
```

SELECT statement chooses team name, year, average salary, and number of peoples to display from the data tables.

Joins two tables: Salaries and Teams so we can match each team’s salary information to the team records using teamID and yearID.

WHERE clause only includes records from the years 2010 to 2015.

AVG calculates the average salary for each team in each year, and COUNT counts the number of peoples for each team and year.

Groups the results by team name and year so each team’s yearly average and people count are together in one row.

HAVING filters the teams to only show those with an average salary above 6,800,000 and more than 20 peoples.

In the final part, it selects each team name, year, average salary, and number of people.

ORDER BY avgSalary DESC sorts teams so those with the highest average salary appear first.

LIMIT 10 lists only the top 10 results that meet all the conditions.

```python
%%sql
SELECT
    t.name AS teamName,
    s.yearID,
    AVG(s.salary) AS avgSalary,
    COUNT(s.playerID) AS numpeoples
FROM Salaries s
JOIN Teams t 
ON s.teamID = t.teamID 
AND s.yearID = t.yearID
WHERE s.yearID between 2010 and 2015
GROUP BY t.name, s.yearID
ORDER BY avgSalary DESC
LIMIT 10;
```

```python
%%sql
WITH payroll AS (
    SELECT
        
        t.name AS teamName,
        s.teamID,
        s.yearID AS Salary_Year,
        SUM(s.salary) AS totalPayroll
       
    FROM Salaries s
    INNER JOIN People p 
    ON s.playerID = p.playerID
    INNER JOIN Teams t 
    ON s.teamID = t.teamID 
    AND s.yearID = t.yearID
    GROUP BY s.teamID, s.yearID
)

Select 
teamID,
teamName, 
Salary_Year, 
totalPayroll
From payroll
group by teamID, teamName, Salary_Year
;
```

SELECT statement chooses specific columns: the team ID, team name, salary year, and total payroll.

WITH statement creates temporary data called payroll. It acts as a short-term table containing all payroll calculations for each team and year so the query is easier to organize.

Joins Salaries, People, and Teams tables so the query combines salary data with player and team details, linked by playerID, teamID, and yearID.

SUM function adds up player salaries to calculate the total payroll for each team in a given year.

GROUP BY in both the WITH clause and final SELECT ensures each row displays the total payroll for every team and year combination.


```python
%%sql
WITH payroll AS (
    SELECT
        p.nameFirst || ' ' || p.nameLast AS playerName,
       
        s.yearID AS Salary_Year,
        
        SUM(t.HR) AS totalHomeRuns
    FROM Salaries s
    INNER JOIN People p 
    ON s.playerID = p.playerID
    INNER JOIN Teams t 
    ON s.teamID = t.teamID 
    AND s.yearID = t.yearID
    GROUP BY s.teamID, s.yearID
)

Select 
playerName,
 
Salary_Year, 
totalHomeRuns
From payroll
group by playerName, Salary_Year
having max(totalHomeRuns)
;
```

WITH statement creates temporary data called payroll. It acts as a short-term table containing each player's name, the year of their salary, and the total home runs for their team and year.

Joins Salaries, People, and Teams tables so the query combines salary data with player and team details, linked by playerID, teamID, and yearID.
