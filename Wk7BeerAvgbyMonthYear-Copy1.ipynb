```python
#Set display option to show all columns - Configures Pandas to display all columns of a DataFrame when printed or 
#displayed in an interactive environment (e.g., Jupyter Notebook)
```

```python
pd.set_option('display.max_columns',None)
#pd.set_option(): A Pandas function used to configure global display options.
#display.max_columns: Specifies the option to control the maximum number of columns displayed.
#None: Sets the option to display all columns, regardless of how many there are
myBeer_Review
```

```python
#Groups the data by the date column (representing date)
#Averages the beer review scores to calculate (mean) the Avg_Score column for each date group to calculate Avg_Score by date
#Sorts the calculated avg score values in ascending order
#Resets the index and renames the calculated column as Avg_Spend
pd.set_option('display.max_rows',None)
myAvgBeer_Review=myBeer_Review.groupby('date')['score'].mean().sort_values().reset_index(name='Avg_Score')
```

```python
# Displays the resulting head of the DataFrame (myAvgBeer_Review) containing the average calculated score values.
```

```python
# Configure pandas to display all rows in output (no truncation with "...")
pd.set_option('display.max_rows',None)
```

```python
# Extract components from the datetime column
myAvgBeer_Review['year'] = pd.to_datetime(myAvgBeer_Review['date']).dt.year
myAvgBeer_Review['month'] = pd.to_datetime(myAvgBeer_Review['date']).dt.month
```

```python
# Group by year and month, calculate average of the already-averaged scores
```

```python
myAvgBeer_Review.groupby(['year','month']).value_counts().sort_index()
```
