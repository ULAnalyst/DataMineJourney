```python
# Step 2 - # read_csv File Data to be loaded into pandas Data Frame (similar to tables in sql) - click run
```

```python
mySports = pd.read_csv("/anvil/projects/tdm/data/olympics/athlete_events.csv")
```

```python
# Step 3 - # Display first 5 rows of data from the athletes events data frame(similar to Top 5 in sql) - click run
```

```python
mySports.head()
```

```python
# Fiter rows where 'Team' is exactly "Denmark" - Filters the mySports DataFrame to include only rows where the Team column has the exact value "Denmark.
```

```python
mySports[mySports['Team'] ==('Denmark')] #mySports['Team'] == ('Denmark'): Creates a boolean mask that is True for rows where the Team value is "Denmark
```

```python
#Count the number of rows where 'Team' is exactly "Denmark" - Counts the number of rows in the filtered DataFrame where the Team is "Denmark
```

```python
len(mySports[mySports['Team'] ==('Denmark')]) -len(...): Returns the number of rows in the filtered DataFrame
```

```python
#Filter rows where 'Team" contains "Denmark but is not exactly Denmark - Filters the mySports DataFrame to include rows where
```

```python
mySports[(mySports['Team'].str.contains('Denmark')) &  
(mySports['Team'] !=('Denmark'))]  #The Team column contains the substring "Denmark.
#The Team column is not exactly equal to "Denmark
```

```python
#Count occurences of each unique team name meeting both conditions - Counts how many times each unique team name appears in rows where
```

```python
len(mySports[(mySports['Team'].str.contains('Denmark')) &  
(mySports['Team'] !=('Denmark'))])
```

```python
#Save filtered and counted data inside of a variable named "mySports_Denmark" -Stores the result of step 5 into a variable named mySports_Denmark
```

```python
mySports_Denmark = mySports[(mySports['Team'].str.contains('Denmark')) &  
(mySports['Team'] !=('Denmark'))]['Team'].value_counts() #This variable can now be used for further analysis or visualization
print(mySports_Denmark) #Print the filtered and counted data
```

```python
#Display Bar Chart - Renders and displays the bar chart on the screen.
```

```python
plt.figure(figsize=(15,5)) # <-- figsize=(15,5): Sets the width to 15 inches and height to 5 inches, ensuring a horizontal layout suitable for bar charts with longer labels.
plt.bar(mySports_Denmark.index, mySports_Denmark, color='teal')
plt.xlabel("Denmark Teams", fontsize=12) #"Name of Denmark Teams": Indicates that each bar corresponds to a Denmark Team in the (National Olympic Committee).
plt.ylabel("Number of Denmark Athletes", fontsize=12) #Number of Denmark Athletes: Indicates that the y-axis represents how many times each Denmark appears in the dataset.
plt.title("Frequency of Denmark Athletes in the Olympics ", fontsize=16) #Frequency of Team Name of Denmark & !(not)= Denmakr : Describes what is being visualized (the frequency/nunber of times Denmark is represented in the dataset).
plt.show() #Display Chart - Without this line, no visualization will appear when running scripts in some environments (e.g., Jupyter Notebook or standalone Python scripts).
```

```python
#Group by 'NOC' and calculate avg Athlete Height - Groups the mySports variable by National Olympic Commission (NOC), 
#calculates the average athlete height (Athlete_Avg_Height) for each NOC, 
#resets the index and renames the averaged column as Athlete_Avg_Height
```

```python
pd.set_option('display.max_rows',None) # - pd.set_option(): A Pandas function used to configure global display options.
# - display.max_columns: Specifies the option to control the maximum number of columns displayed.
# - None: Sets the option to display all columns, regardless of how many there are
mySports.groupby('NOC')['Height'].mean().reset_index(name='Athlete_Avg_Height')
```

```python
#Group by 'Sport' and calculate avg Athlete Height - Groups the mySports variable by Sport, 
#calculates the average athlete height (Athlete_Avg_Height) for each Sport, 
#resets the index and renames the averaged column as SportsAthlete_Avg_Height
```

```python
pd.set_option('display.max_rows',None) # - pd.set_option(): A Pandas function used to configure global display options.
# - display.max_columns: Specifies the option to control the maximum number of columns displayed.
# - None: Sets the option to display all columns, regardless of how many there are
mySports.groupby('Sport')['Height'].mean().reset_index(name='SportsAthlete_Avg_Height')
```
