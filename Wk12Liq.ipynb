```python
# Load sales data with only the necessary columns and optimized data types for analysis
```

```python
# Loads the full dataset but only the columns needed for sales analysis, optimizing memory usage and preserving zip code formatting
myLiqSales =pd.read_csv("/anvil/projects/tdm/data/iowa_liquor_sales/iowa_liquor_sales.csv", usecols=["Zip Code", "Category Name", "Sale (Dollars)"], dtype={"Zip Code": "string", "Category Name": "string", "Sale (Dollars)": "float"})
```

```python
# Display the loaded sales DataFrame to verify correct columns and data types
myLiqSales
```

```python
# Group by Zip Code, sum Sale (Dollars), and reset index for easy sorting and display
```

```python
# Aggregates total sales for each zip code, preparing the data for identifying top-performing locations
top_zip_sales = (myLiqSales.groupby('Zip Code')["Sale (Dollars)"].agg('sum').reset_index())
top_zip_sales.nlargest(10, 'Sale (Dollars)') # Finds and displays the 10 zip codes with the highest total sales, providing insight into the higheast Sale (Dollars)
```

```python
# Group by Category Name, sum Sale (Dollars), and reset index for easy sorting and display
```

```python
# Aggregates total sales for each product category, enabling analysis of product performance across the state
top_categ_sales = (myLiqSales.groupby('Category Name')["Sale (Dollars)"].agg('sum').reset_index())
top_categ_sales.nlargest(10, 'Sale (Dollars)') # Finds and displays the 10 product categories with the highest total sales, revealing top-selling liquor categories
```
