```python
# Step 1 - Import Required Library -The pandas library is imported(the pd is an alias for pandas - similar to database in sql) - click run
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Step 2 - # read_csv File Data to be loaded into pandas Data Frame (similar to tables in sql) - click runmyGroceries = pd.read_csv("/anvil/projects/tdm/data/8451/The_Complete_Journey_2_Master/5000_transactions.csv")
myGroceries = pd.read_csv("/anvil/projects/tdm/data/8451/The_Complete_Journey_2_Master/5000_transactions.csv")
# Step 3 - # Display first 5 rows of data from the Grocery Store Transaction data frame(similar to Top 5 in sql) - click run
# Assign Column Names - Renames the columns of the myGroceries DataFrame to
# ensure that column names are easy to work with and follow a standardized format.
myGroceries.columns = ["BASKET_NUM", "HSHD_NUM", "PURCHASE", "PRODUCT_NUM", "SPEND", "UNITS", "STORE_R", "WEEK_NUM", "YEAR"]
# Step 4 - # Display first 5 rows of data from the Gorcery Store Transaction data frame(similar to Top 5 in sql) - click run
myGroceries.head()
#Set display option to show all columns - Configures Pandas to display all columns of a DataFrame when printed or 
#displayed in an interactive environment (e.g., Jupyter Notebook)
pd.set_option('display.max_columns',None)
#pd.set_option(): A Pandas function used to configure global display options.
#display.max_columns: Specifies the option to control the maximum number of columns displayed.
#None: Sets the option to display all columns, regardless of how many there are
myGroceries
#Group by 'YEAR' and calculate sum/"Total"'Spend' - Groups the grocery store transaction DataFrame by transaction (YEAR), 
#calculates the sum (Total_Spend) for each YEAR, and resets the index and renames the aggregated column as Total_Spend
pd.set_option('display.max_rows',None)
myGroceries.groupby('YEAR')['SPEND'].sum().reset_index(name='Total_Spend')

#Group by 'STORE_R' and calculate sum/"Total"'Spend' - Groups the grocery store transaction DataFrame by store region (STORE_R), 
#calculates the sum (Total_Spend) for each STORE_R, and resets the index and renames the aggregated column as Total_Spend
#pd.set_option('display.max_rows',None)
myGroceries.groupby('STORE_R')['SPEND'].sum().reset_index(name='Total_Spend')
```
