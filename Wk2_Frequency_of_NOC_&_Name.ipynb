```python
#Counts the number of occurrences of each unique value in the NOC column of the mySports DataFrame.
```

```python
countOf_NOC = mySports['NOC'].value_counts(dropna = False)
print(countOf_NOC) # <--print all "NOC" counts - Displays the full list of NOCs and their corresponding counts in descending order of frequency
```

```python
# Selects the top 10 most frequent NOCs from the countOf_NOC Series.
```

```python
top10_NOC = countOf_NOC.head(10)
print(top10_NOC) # <-- print the top 10 NOCs - Displays the top 10 NOCs and their counts for verification.
```

```python
# Create Figure for plotting graph - Initializes a new figure for plotting with a specified size.
```

```python
plt.figure(figsize=(15,5)) 
```

```python
# Plot bar chart - plt.bar(top10_NOC.index, top10_NOC)
```

```python
plt.bar(top10_NOC.index, top10_NOC)
```

```python
# Add x axis label - Adds a label to the x-axis to describe what it represents.
```

```python
plt.xlabel("National Olympic Committee", fontsize=12) 
```

```python
#add y axis label - Adds a label to the y-axis to describe what it represents.
```

```python
plt.ylabel("Count of NOC", fontsize=12) 
```

```python
# Add title to chart - Adds a descriptive title to the bar chart.
```

```python
plt.title("Frequency of National Olympic Committee", fontsize=16) 
```

```python
#Display Bar Chart - Renders and displays the bar chart on the screen.
```

```python
plt.figure(figsize=(15,5)) # <-- figsize=(15,5): Sets the width to 15 inches and height to 5 inches, ensuring a horizontal layout suitable for bar charts with longer labels.
plt.bar(top10_NOC.index, top10_NOC, color='purple')
plt.xlabel("National Olympic Committee", fontsize=12) #"Name of Country": Indicates that each bar corresponds to an NOC (National Olympic Committee), which is often associated with a country or region.mpic Committee), which is often associated with a country or region.
plt.ylabel("Count of NOC", fontsize=12) #Count of NOC: Indicates that the y-axis represents how many times each NOC appears in the dataset.
plt.title("Frequency of National Olympic Committee", fontsize=16) #Frequency of Name of Committee: Describes what is being visualized (the frequency/counts of countries or regions represented by their NOCs).
plt.show() #Display Chart - Without this line, no visualization will appear when running scripts in some environments (e.g., Jupyter Notebook or standalone Python scripts).
```

```python
#Counts the number of occurrences of each unique value in the Name column of the mySports DataFrame.
```

```python
countOf_Name = mySports['Name'].value_counts(dropna = False)
print(countOf_Name)
```

```python
# Selects the top 10 most frequent Name's from the countOf_Name Series.
```

```python
top10_Name = countOf_Name.head(10)
print(top10_Name)
```

```python
#Display Bar Chart - Renders and displays the bar chart on the screen.
```

```python
plt.figure(figsize=(12,8))
plt.barh(top10_Name.index, top10_Name)
plt.xlabel("Number of Times Name appears", fontsize=12)
plt.ylabel("Athlete Names", fontsize=12)
plt.title("Frequency of Athelete Names", fontsize=16)
plt.show()
```
