```python
#Set display option to show all columns - Configures Pandas to display all columns of a DataFrame when printed or 
#displayed in an interactive environment (e.g., Jupyter Notebook)
```

```python
pd.set_option('display.max_columns',None)
#pd.set_option(): A Pandas function used to configure global display options.
#display.max_columns: Specifies the option to control the maximum number of columns displayed.
#None: Sets the option to display all columns, regardless of how many there are
myGroceries
```

```python
#Group by 'PRODUCT_NUM' and calculate sum "Total"'Spend' - Groups the grocery store product transactions DataFrame by product (PRODUCT_NUM), 
#calculates the sum (Total_Spend) for each PRODUCT_NUM, sort the values, call the last 10 rows of data, and resets the index and renames the aggregated column as Total_Spend
```

```python
pd.set_option('display.max_rows',None)
myTop10_Prod=myGroceries.groupby('PRODUCT_NUM')['SPEND'].sum().sort_values().tail(10).reset_index(name='Total_Spend')
#Groups the data by the PRODUCT_NUM column (representing product numbers)
#Aggregates (sums) the SPEND column for each product group to calculate total spend per product
#Sorts the aggregated total spend values in ascending order
#Selects the top 10 products with the highest total spend by taking the last 10 rows of the sorted data
#Resets the index and renames the aggregated column as Total_Spend

```

```python
# Displays the resulting DataFrame (myTop10_Prod) containing the top 10 products with there total spend values.
```

```python
myTop10_Prod
```

```python
# Sorts the myTop10_Prod DataFrame by the Total_Spend column in descending order (ascending=False) to display products with the highest spend at the top.
```

```python
myTop10_Prod.sort_values(by='Total_Spend',ascending=False)
```
