```python
# Assign Column Names - Renames the columns of the myGroceries DataFrame to
# ensure that column names are easy to work with and follow a standardized format.
```

```python
myGroceries.columns = ["BASKET_NUM", "HSHD_NUM", "PURCHASE", "PRODUCT_NUM", "SPEND", "UNITS", "STORE_R", "WEEK_NUM", "YEAR"]
```

```python
#Set display option to show all columns - Configures Pandas to display all columns of a DataFrame when printed or 
#displayed in an interactive environment (e.g., Jupyter Notebook)
```

```python
pd.set_option('display.max_columns',None)
#pd.set_option(): A Pandas function used to configure global display options.
#display.max_columns: Specifies the option to control the maximum number of columns displayed.
#None: Sets the option to display all columns, regardless of how many there are
myGroceries
```

```python
# Group by YEAR and STORE_R, calculate sum of SPEND for each group
# Then sort by YEAR in ascending order (as requested)
# The 'name' parameter renames the SPEND sum column to 'Total_Spend'
pd.set_option('display.max_rows',None)
myAnnualSpend=myGroceries.groupby(['YEAR','STORE_R'])['SPEND'].sum().sort_values().reset_index(name='Total_Spend').sort_values('YEAR')
```
